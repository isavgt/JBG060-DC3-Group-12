```python
import pandas as pd
import random
```

```python
df_summaries = pd.read_csv("data/articles_senti_topics.csv")
df_pars = pd.read_csv("data/all_africa_southsudan.csv")
text = df_pars.loc[11124, "paragraphs"]
```

Get 100 random indexes of summaries and paragraphs and divide them between the group members

```python
random.seed(5102023)

amount_of_checks = 60
random_summaries = ([random.randint(0, len(df_summaries)) for i in range(amount_of_checks)])
random_pars = ([random.randint(0, len(df_pars)) for i in range(amount_of_checks)])

part_size = amount_of_checks // 3

summaries_hilde_isa = sorted(random_summaries[:part_size])
summaries_evert_carlijn = sorted(random_summaries[part_size:2 * part_size])
summaries_giulia_dani = sorted(random_summaries[2 * part_size:])

pars_hilde_isa = sorted(random_pars[:part_size])
pars_evert_carlijn = sorted(random_pars[part_size:2 * part_size])
pars_giulia_dani = sorted(random_pars[2 * part_size:])

division_data = {
    'summaries_hilde_isa': summaries_hilde_isa,
    'summaries_evert_carlijn': summaries_evert_carlijn,
    'summaries_giulia_dani': summaries_giulia_dani,
    'pars_hilde_isa': pars_hilde_isa,
    'pars_evert_carlijn': pars_evert_carlijn,
    'pars_giulia_dani': pars_giulia_dani
}

df_division = pd.DataFrame(division_data)
```

Get the correct slices of the full dataframe and write them into new csvs

```python
random_summaries_limited = df_summaries[df_summaries.index.isin(random_summaries)][["summary"]]
random_pars_limited = df_pars[df_pars.index.isin(random_pars)]

random_summaries_limited.to_csv("data/random_summaries.csv")
random_pars_limited.to_csv("data/random_pars.csv")
#random_pars_limited
```
